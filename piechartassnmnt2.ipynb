```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv('computer_parts.csv')
```

```python
product = df["Product Name;"]
bugs = df["Number Of Bugs"]
colors = ["orange", "red", "green", "skyblue", "yellow"]
```

```python
#prediction without adjustment

plt.pie(bugs, labels=product, colors=colors, autopct='%.2f%%', shadow=True)
plt.show()
```

```python
#after concatinate or sum of similar product name

conc = df.groupby(df["Product Name;"])["Number Of Bugs"].sum()
plt.pie(conc, labels=conc.index, colors=colors, autopct='%.2f%%', shadow=True)
plt.axis('equal');
plt.show()
```
