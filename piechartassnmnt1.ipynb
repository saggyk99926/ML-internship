```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv('dataset.csv')
```

```python
country = df["country"]
medal = df["gold_medal"]
colors = ["orange", "red", "green", "skyblue", "yellow"]
```

```python
plt.pie(medal, labels=country, colors=colors, autopct='%.2f%%', shadow=True)
plt.legend(country)
plt.axis('equal')
plt.show()
```
