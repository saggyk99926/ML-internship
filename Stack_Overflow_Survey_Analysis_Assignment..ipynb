```python
#Which  is most used  programming langauge (use LanguageWorkedWith column)
count=df["LanguageWorkedWith"].value_counts() 
count.head(1)
```

```python
#Display how many work remotely on pie chart  [use WorkRemote column]
import matplotlib.pyplot as plt
#group_df=df.groupby('WorkRemote')
group=df['WorkRemote'].value_counts()
plt.pie(group,labels=group,autopct='%.1f%%')
plt.legend(group,loc="upper right")
plt.axis('equal')
plt.show()
group
```

```python
#Display how many male and female programmers on pie chart  [use Gender column]
gender=df['Gender'].value_counts()
result=gender.head(2)
label=['Men','Female']
plt.pie(result,labels=label,autopct="%.1f%%",shadow=True)
plt.legend(label)
plt.axis('equal')
plt.show()
```
